```python
import requests
from bs4 import BeautifulSoup
```

```python
# open the url and get the page
url='https://www.puppyleaks.com/popular-dog-names'
r=requests.get(url)
r.encoding = 'UTF-8'

#Write the page to a intermediate file
with open("dognames.html", mode='wb') as targetFile:
    targetFile.write(r.content)
```

```python
# open the webpage
with open('dognames.html',mode='r',encoding="utf8") as targetFile:
    html_content=targetFile.read()
```

```python
soup = BeautifulSoup(html_content,'html.parser')
dognames=[]
df_list=soup.find_all(class_="column-1")
for _ in range(1,len(df_list)):
    Text=str(df_list[_])
    dognames.append(Text.split('>',3)[1].split('<')[0])
    
df_list=soup.find_all(class_="column-2")
for _ in range(1,len(df_list)):
    Text=str(df_list[_])
    dognames.append(Text.split('>',3)[1].split('<')[0])
    
```
